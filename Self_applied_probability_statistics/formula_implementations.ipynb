```python
import scipy as sc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Binomial Distribution, PMF formula

```python
def binomial(n,r,p):
    """
    n: number of trials
    r: number of successful events
    p: probabiliy of successful event
    """
    # ncr = sc.special.factorial(n)/(sc.special.factorial(n-r)*sc.special.factorial(r))
    ncr = sc.special.comb(n,r) #Scipy already gives permuations and combinations
    pr = p**r
    qr = (1-p)**(n-r)
    px = ncr*pr*qr
    # print(px)
    return px
```

```python
binomial(5,3,1/6)
```

    Creating binomial distribution

```python
def binomial_distribution(n, p):
    x = np.arange(0, n+1)
    probabilities = []
    for r in x:
        probabilities.append(binomial(n, r, p))
    df = pd.DataFrame({
        "x": x,
        "Probability": probabilities
    })
    return df


distribution = binomial_distribution(10, 0.5)
distribution.head(3)
```

```python
distribution.plot(
    x="x",
    y="Probability",
    kind="bar"
)
```
